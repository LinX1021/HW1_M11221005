# 匯入套件及資料集

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

col_name = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
            'marital-status', 'occupation', 'relationship', 'race', 'sex',
            'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']

ori_data = pd.read_csv(r'C:\Users\M11021001\OneDrive - 國立雲林科技大學\Desktop\資料探勘\作業1\adult\adult.data', header=None)

# 更改欄位名稱
ori_data.columns = col_name
ori_data
```

# 檢查空值(空值會用問號表示)

```python
ori_data.apply(lambda x:np.sum(x==' ?'))
```

```python
# 建立規則，目的是要撈出沒有缺失值的資料(即刪除缺失值資料)
rule1 = ori_data['workclass'] != ' ?'
rule2 = ori_data['occupation'] != ' ?'
rule3 = ori_data['native-country'] != ' ?'

df = ori_data[(rule1 & rule2 & rule3)]
df
```

# 再次檢查有無空值

```python
df.apply(lambda x:np.sum(x==' ?'))
```

# 刪除對結果無影響的欄位

```python
df = df.drop(labels=['fnlwgt'],axis=1)
df
```

# 找到資料型態不是數值的欄位

```python
df.info()
```

# 將類別資料轉數值

```python
from sklearn.preprocessing import LabelEncoder

# 用.info()觀察到的類別欄位如下
obj_col = ['workclass', 'education', 'marital-status', 'occupation',
           'relationship', 'race', 'sex', 'native-country', 'income']

encoder = LabelEncoder()
data_le = pd.DataFrame(df)

# 將類別欄位進行編碼
for col in data_le[obj_col]:
    data_le[col] = encoder.fit_transform(data_le[col])
data_le
```

```python
count0 = 0
for o in data_le['income']:
    if o == 0:
        count0 += 1
print(count0)

count1 = 0
for o in data_le['income']:
    if o == 1:
        count1 += 1
print(count1)
```

```python
col_name_after = data_le.columns
len(col_name_after)
```

```python
data_le.to_csv(r'C:\Users\M11021001\OneDrive - 國立雲林科技大學\Desktop\資料探勘\作業1\adult_clean_data_ori.csv')
```

# 資料正規化

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler(feature_range=(0,1))
data_nor = min_max_scaler.fit_transform(data_le)
data_nor = pd.DataFrame(data_nor)
data_nor.columns = col_name_after
data_nor
```

# 完成資料前處理，另存新檔供後續使用

```python
data_nor.to_csv(r'C:\Users\M11021001\OneDrive - 國立雲林科技大學\Desktop\資料探勘\作業1\adult_clean_data.csv')
```
